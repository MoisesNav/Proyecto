```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/INEGI_DENUE_20022024.csv', encoding='latin-1')
```

```python
# Configurar para mostrar todas las columnas
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
# Obtener el listado de columnas
listado_columnas = df.columns

# Mostrar el listado de columnas
print(listado_columnas)
```

```python
# Seleccionar solo las columnas deseadas
columnas_deseadas = ['Código Postal', 'Localidad', 'Tipo de asentamiento humano', 'Nombre de asentamiento humano']
df_seleccionado = df[columnas_deseadas]


df_seleccionado.head()
```

```python
# Eliminar duplicados basados en la columna 'Código Postal'
df_sin_duplicados = df_seleccionado.drop_duplicates(subset=['Código Postal'])
df_sin_duplicados
```

```python
df_sin_nulos = df_sin_duplicados.dropna()
df_sin_nulos
```

```python

# Ordenar el DataFrame por la columna 'Nombre de asentamiento humano'
df_ordenado = df_sin_nulos.sort_values(by='Nombre de asentamiento humano')
df_ordenado
```

```python
# Guardar el DataFrame ordenado como un nuevo archivo CSV
df_ordenado.to_csv('/content/INEGI_DENUE_ordenado.csv', index=False)

# Verificar que el nuevo archivo se haya creado correctamente
print("DataFrame ordenado guardado como INEGI_DENUE_ordenado.csv")
```

```python

# Eliminar filas duplicadas basadas en la columna 'Nombre de asentamiento humano'
df_sin_duplicadosH = df_ordenado.drop_duplicates(subset=['Nombre de asentamiento humano'])

# Ordenar el DataFrame resultante por la columna 'Nombre de asentamiento humano'
df_sin_duplicados_ordenadoh = df_sin_duplicadosH.sort_values(by='Nombre de asentamiento humano')

# Guardar el DataFrame sin duplicados y ordenado como un nuevo archivo CSV
df_sin_duplicados_ordenadoh.to_csv('/content/INEGI_DENUE_sin_duplicados_ordenado.csv', index=False)

# Verificar que el nuevo archivo se haya creado correctamente
print("DataFrame sin duplicados y ordenado guardado como INEGI_DENUE_sin_duplicados_ordenado.csv")
```
